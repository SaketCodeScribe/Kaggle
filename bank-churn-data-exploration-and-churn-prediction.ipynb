```python
data['Attrition_Flag'].unique()
```

<h3>Attrition Flag is a categorical variable</h3>

```python
ex.pie(data, names='Attrition_Flag', title='Proportion of Customer existing')
```

<h3>We can see that our target class is unbalanced.</h3>

<h3>As a general rule of thumb we consider Customers Age as a numerical vairable.</h3>

```python
fig=mk_sp(rows=2, cols=1)

x1=go.Box(x=data['Customer_Age'], name='Age Box Plot', boxmean=True)
x2=go.Histogram(x=data['Customer_Age'], name='Age Histogram')

fig.add_trace(x1, row=1, col=1)
fig.add_trace(x2, row=2, col=1)

fig.update_layout(height=700, width=1200, title_text="Distribution of Customers Ages")
fig.show()

```

<h3>lower fence = Q1 - 1.5*IQR = 24.5, upper fence = Q3 + 1.5*IQR = 68 (after rounding) and thus we get two ouliers one for age=70 and another for age=73 (univariate outlier).<br><br>
We can see that the distribution of customer's age follows a fairly normal distribution</h3>

```python
data['Dependent_count'].unique()
```

<h3>Since there are only 5 unique values we can take it as a categorical variable</h3>

```python
ex.pie(data, names='Dependent_count', title='Proportion of People Financially Dependent on Customers')
```

<h3>We can see that our data has the highes percentage of customers who has 3 financial dependent persons and lowest percentage of customers who has 5 financial dependent persons.<br></h3>
<h3 style="background-color:lightgreen">But, this attribute's value is not fixed as logically there can be values > 5, so for better generalization of our model we'll consider Dependent Count as a numerical variable.</h3>

```python
fig=mk_sp(rows=2, cols=1)

x1=go.Box(x=data['Dependent_count'], name='Age Box Plot', boxmean=True)
x2=go.Histogram(x=data['Dependent_count'], name='Age Histogram')

fig.add_trace(x1, row=1, col=1)
fig.add_trace(x2, row=2, col=1)

fig.update_layout(height=700, width=1200, title_text="Distribution of People Financially Dependent on Customers")
fig.show()
```

<h3>We can see that the distribution of dependent count follows a normal distribution</h3>
